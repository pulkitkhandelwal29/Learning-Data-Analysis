```python
Pandas Built-in Visualisation--They are just helpful in quick analysis, to customize we need to use seaborn
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline

df1=pd.read_csv('df1.csv',index_col=0)					//it reads the csv and apply index as the time series
df2=pd.read_csv('df2')

//histogram is the most commonly used when we make 
df1['A'].hist()					//creates a histogram,calling matplotlib under the hood
df1['A'].plot(kind='hist',bins=30)		//creates a histogram by another method
df1['A'].plot.hist()		//used method


df2.plot.area(alpha=0.4)		//displays an area plot with more transparency because of alpha parameter


df2.plot.bar(stacked=True)				//creates a bar plot and in this index should be less and this displays the chart same as in combining data to make chart with different colors in the same bar


df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)		//displays line plot with x axis values and y axis values

***All the paramters we learnd in matplotlib, can be applied to these pandas dataframes



df1.plot.scatter(x='A',y='B')			//creates a scatter plot


df2.plot.box()			//boxplot


df=pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')			//hexabin plots


df2.plot.kde()					//just line but not histogram,kdeplotss















Python for data visualisation-Plotly and Cufflinks
Plotly is an interactive visualisation library
Cufflinks connects plotly with pandas

import numpy as np
import pandas as pd
%matplotlib inline

import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot

init_notebook_mode(connected=True)				//connects the javascript to the notebook
cf.go_offline()								//helps cufflinks use offline


df=pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())		//create a dataframe with the data on random
df2=pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})


df.plot()				//used to create plot as normal

df.iplot()				//creates an interative plot 


df.iplot(kind='scatter',x='A',y='B',mode='markers',size=20)			//creates a scatter plot with markers on it, not the conencted lines and as it is scatter we need to pass x and y values


df2.iplot(kind='bar',x='Category',y='Values')		//creates a bar chart


df.iplot(kind='box')						//boxplot




df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[5,4,3,2,1]})
df3.iplot(kind='surface',colorscale='rdylbu')					//creates a 3D surface plot



df['A'].iplot(kind='hist',bins=50)					//histogram(bins make it smaller and understandable)
df.iplot(kind='hist')			//we can pass the whole dataframe and later in interactive plot we can switch it off and on


df[['A','B']].iplot(kind='spread')						//spread is a type of plot used mainly in stock market



df.iplot(kind='bubble',x='A',y='B')				//bubble plots


df.scatter_matrix()					//not good for the large datasets














Python for Data Visualization - Geographical Plotting
It is more challenging to due the various formats the data can come in
Plotly can be used for plotting
Matplotlib also has a basemap extension


Choropleth Maps-
It helps to plot on a global scale or nation wide scale or world scale

import plotly.plotly as py
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)

data=dict(type=choropleth,locations=['AZ','CA','NY'],locationmode='USA-states',colorscale='Portland',text=['text 1','text 2','text 3'],z=[1.0,2.0,3.0],colorbar={'title':'Colorbar title goes here'})

//locations are the real places we need to place,locationmode is the area where we want to plot(given in the documentation), text is the name of the locations and z is the value which we want to fill in there


layout=dict(geo={'scope':'usa'})
choromap=go.Figure(data=[data],layout=layout
iplot(choromap)						//plotting at the usa country




>>We need to reference documentation to make it better, u can't always remember<<
We need to explore more and do copy paste from the documentation
Feed from Live data-
import pandas as pd
df=pd.read_csv('2011_US_AGRI_Exports')

data=dict(type=choropleth,locations=df['code'],locationmode='USA-states',colorscale='Portland',text=df['text'],z=df['total exports'],colorbar={'title':'Millions USD'},marker=dict(line=dict(color='rgb(255,255,255)',width=2))

//marker is used as the spacing between the states

layout=dict(title='2011 US Agriculture Exports by state, geo=dict(scope='usa',showlakes=True,lakecolor='rgb(85,173,240)'))

choromap2=go.Figure(data=[data],layout=layout)
iplot(choromap2)




Choropleth International maps-
df=pd.read_csv('2014_World_GDP')

data=dict(type=choropleth,locations=df['code'],text=df['country'],z=df['GDP (Billions)'],colorbar={'title':'GDP in Billions USD'})

layout=dict(title='2014 Global GDP, geo=dict(showframe=False,projection={'type':'Mercator'}))

choromap3=go.Figure(data=[data],layout=layout)
iplot(choromap3)
```
